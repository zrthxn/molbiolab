```r
library("patchwork")
data = read.csv("data/tempuf_30-36_MasineAli.csv")
```

```r
colnames(data)
```

# Lysozyme

```r
# Ratio - 350nm : 330nm
MAXLYS = max(data$Ratio.350.nm...330.nm.for.Cap.34)

plot(x = data$Temperature.for.Cap.34...C., y = data$Ratio.350.nm...330.nm.for.Cap.34/MAXLYS, col="orange",
     type = "l", 
     ylim = c(0.75,1), 
     xlab="Temperature", ylab="Ratio - 350nm : 330nm",
     main="Lysozyme Temperature Spectroscopy")
lines(x = data$Temperature.for.Cap.35...C., y = data$Ratio.350.nm...330.nm.for.Cap.35/MAXLYS, col="red") 
lines(x = data$Temperature.for.Cap.36...C., y = data$Ratio.350.nm...330.nm.for.Cap.36/MAXLYS, col="pink")
```

```r
# Emission of 350nm

Eg1 <- plot(x = data$Temperature.for.Cap.34...C., y = data$X350.nm.for.Cap.34, col="orange",
     type = "l", 
     ylim = c(200,3200),
     xlab="Temperature", ylab="Emission of 350nm",
     main="Lysozyme 350nm Emmission") +
lines(x = data$Temperature.for.Cap.35...C., y = data$X350.nm.for.Cap.35, col="red") +
lines(x = data$Temperature.for.Cap.36...C., y = data$X350.nm.for.Cap.36, col="pink")

# Emission of 330nm

Eg2 <- plot(x = data$Temperature.for.Cap.34...C., y = data$X330.nm.for.Cap.34, col="orange",
     type = "l", 
     ylim = c(200,3200),
     xlab="Temperature", ylab="Emission of 330nm",
     main="Lysozyme 330nm Emmission") + 
lines(x = data$Temperature.for.Cap.35...C., y = data$X330.nm.for.Cap.35, col="red") +
lines(x = data$Temperature.for.Cap.36...C., y = data$X330.nm.for.Cap.36, col="pink")

Eg1 + Eg2
```
