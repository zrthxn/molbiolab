# Kinetics
Michaelis-Menten enzyme kinetics of Pyruvate to Lactate reaction.

$Pyruvate +  NADH + LDH -> Lactate$

## Original Files

```r
fnames = list.files("data/Michaelis-Menten original")
fnames
```

```r
junk <- c(fnames[6], fnames[7], fnames[9])
baselines <- c(fnames[1], fnames[18], fnames[19], fnames[20])
order <- c(
    fnames[8],
    fnames[10], 
    fnames[11], 
    fnames[12], 
    fnames[13], 
    fnames[14], 
    fnames[15], 
    fnames[16],
    fnames[17],
    fnames[5],
    fnames[4],
    fnames[3],
    fnames[2]
)
```

```r
for (f in baselines) {
    df <- read.csv(
        paste("data/Michaelis-Menten original/", f, sep = ""),
        sep=";", dec=",")
    plot(x=df$X.sec., y=df$X1..A., type="l", main=f)
}
```

```r
collated_df <- data.frame(idx = c(1:60))
```

```r
for (i in c(1:length(order))) {
    df <- read.csv(
        paste("data/Michaelis-Menten original/", order[i], sep = ""),
        sep=";", dec=",")
    collated_df[paste('T', i, sep="")] = df$X.sec. 
    collated_df[paste('R', i, sep="")] = df$X1..A.
}
```

```r
df <- read.csv(
    paste("data/Michaelis-Menten original/", baselines[1], sep = ""),
    sep=";", dec=",")
collated_df$BT_ReactionMix = df$X.sec. 
collated_df$BR_ReactionMix = df$X1..A.

df <- read.csv(
    paste("data/Michaelis-Menten original/", baselines[2], sep = ""),
    sep=";", dec=",")
collated_df$BT_RxnM_PBS_1 = df$X.sec. 
collated_df$BR_RxnM_PBS_1 = df$X1..A.

df <- read.csv(
    paste("data/Michaelis-Menten original/", baselines[3], sep = ""),
    sep=";", dec=",")
collated_df$BT_RxnM_PBS_2 = df$X.sec. 
collated_df$BR_RxnM_PBS_2 = df$X1..A.

df <- read.csv(
    paste("data/Michaelis-Menten original/", baselines[4], sep = ""),
    sep=";", dec=",")
collated_df$BT_PBS = df$X.sec. 
collated_df$BR_PBS = df$X1..A.
```

```r
collated_df
```

```r
write.csv(collated_df, "data/kinetics_MasineAli.csv")
```
