$Absorbance = \epsilon . c . d$

$\epsilon = 6220 M^{-1} cm^{-1}$

Plot - rate of change of absorbance of NADH over time, vs, concentration.

From this you can determine the $V_{max}/2$ and so you can get $K_m$, the Michealis-Menten constant.

Find out why the convertion of Py -> Lactate is slowed down at high concentration of the substrate (enzyme NADH)

```r
# Pyruvate mM concentration
concs <- c(5, 2.5, 1.25, 0.625, 0.313, 0.156, 0.078, 0.039, 0.020, 0.010, 0.005, 0.002, 0.001)

# Rate of Reaction (Delta_Abs / Delta_Time)
rates <- c(
    (data$R1[1] - data$R1[60]) / (data$T1[60] - data$T1[1]),
    (data$R2[1] - data$R2[60]) / (data$T2[60] - data$T2[1]),
    (data$R3[1] - data$R3[60]) / (data$T3[60] - data$T3[1]),
    (data$R4[1] - data$R4[60]) / (data$T4[60] - data$T4[1]),
    (data$R5[1] - data$R5[60]) / (data$T5[60] - data$T5[1]),
    (data$R6[1] - data$R6[60]) / (data$T6[60] - data$T6[1]),
    (data$R7[1] - data$R7[60]) / (data$T7[60] - data$T7[1]),
    (data$R8[1] - data$R8[60]) / (data$T8[60] - data$T8[1]),
    (data$R9[1] - data$R9[60]) / (data$T9[60] - data$T9[1]),
    (data$R10[1] - data$R10[60]) / (data$T10[60] - data$T10[1]),
    (data$R11[1] - data$R11[60]) / (data$T11[60] - data$T11[1]),
    (data$R12[1] - data$R12[60]) / (data$T12[60] - data$T12[1]),
    (data$R13[1] - data$R13[60]) / (data$T13[60] - data$T13[1])
)
```

## Analysis to find $K_m$
Fit a linear model to the data, the model estimates will give us an approximation of the constants.

$ \frac{1}{V_i} = \frac{K_m}{V_{max}}\frac{1}{|S|} + \frac{1}{V_{max}} $

$ y = \beta_1x + \beta_0 $

$ V_{max} = \frac{1}{\beta_0} $
$ K_m = \frac{\beta_1}{\beta_0}  $

```r
m <- lm(y ~ x, data=data.frame(y=1/rates, x=1/concs))
summary(m)
```

```r
vmax = 1 / m$coefficients[1]
km = m$coefficients[2] / m$coefficients[1]
print(paste("Vmax =", vmax))
print(paste("Km =", km))
```

```r
plot(y=rates, x=concs, main="Michealis-Menten Plot")
lines(y=rates, x=concs)
abline(h=vmax/2, col="red", lty=2)
text(x=4.1, y=0.0063, labels="50% of Vmax = 0.0058 µM/s", cex=1,col="red") 
abline(v=km-0.05, col="red", lty=2)
text(x=0.80, y=0, labels="Km = 0.1942 µM", cex=1,col="red") 
dev.print(png, 'resources/kinetics_mmplot.png', width=500, height=500)
```

```r
plot(y=1/rates, x=1/concs, main="Lineweaver-Burk Plot")
lines(y=1/rates, x=1/concs, lty=2)
abline(a = m$coefficients[1], b = m$coefficients[2], col="red", lty=1)
legend(0,17000, legend=c("MM Equation", "Data"), col=c("red", "black"), lty=1:2, cex=0.8)
dev.print(png, 'resources/kinetics_lbplot.png', width=500, height=500)
```
